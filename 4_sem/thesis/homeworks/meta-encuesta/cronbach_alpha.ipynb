```python
import pandas as pd
from pingouin import cronbach_alpha
```

```python
data = pd.read_csv("Meta-encuesta.csv")
```

```python
data.head(10)
```

```python
data.dtypes
```

```python
numeric_df = pd.DataFrame(
    columns=[
        'trustable', 
        'max_time_interest', 
        'max_time_no_interest', 
        'lie_shame', 
        'lie_unknown', 
        'lie_privacy', 
        'lie_chaos', 
        'lie_untrusty', 
        'me_untrusty', 
        'lie_speed', 
        'speed_now', 
        '5_min', 
        'lie_here', 
        'share', 
        'chonk'
    ]
)
```

```python
numeric_df['trustable'] = pd.Categorical(data.iloc[:,1]).codes
numeric_df['max_time_interest'] = pd.Categorical(data.iloc[:,2]).codes
numeric_df['max_time_no_interest'] = pd.Categorical(data.iloc[:,3]).codes
numeric_df['lie_shame'] = pd.Categorical(data.iloc[:,4]).codes
numeric_df['lie_unknown'] = pd.Categorical(data.iloc[:,5]).codes
numeric_df['lie_privacy'] = pd.Categorical(data.iloc[:,6]).codes
numeric_df['lie_chaos'] = pd.Categorical(data.iloc[:,7]).codes
numeric_df['lie_untrusty'] = pd.Categorical(data.iloc[:,8]).codes
numeric_df['me_untrusty'] = pd.Categorical(data.iloc[:,9]).codes
numeric_df['lie_speed'] = pd.Categorical(data.iloc[:,10]).codes
numeric_df['speed_now'] = pd.Categorical(data.iloc[:,11]).codes
numeric_df['5_min'] = pd.Categorical(data.iloc[:,12]).codes
numeric_df['lie_here'] = pd.Categorical(data.iloc[:,13]).codes
numeric_df['share'] = pd.Categorical(data.iloc[:,14]).codes
numeric_df['chonk'] = pd.Categorical(data.iloc[:,15]).codes
```

```python
numeric_df.dtypes
```

```python
# Calcular el Alfa de Cronbach
alpha, _ = cronbach_alpha(numeric_df)
print(f'Alfa de Cronbach: {alpha}')
```
