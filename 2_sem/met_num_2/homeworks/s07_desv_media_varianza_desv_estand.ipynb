## Conociendo la base de datos, exportar pandas

```python
import pandas as pd
df = pd.read_csv("s07_titanic_data.csv")
print(df.shape)
df.head()
```

Obtener la desviación Media Absoluta de la columna **Age** de manera manual, posterirmente obtenerla con la funcion de Phyton y comparar resultados

```python
# Desviación media absoluta de columna Age (manual)
age_data = df[["Age"]].copy()
print(age_data.shape)

age_mean = age_data.mean()[0]
print(f"Edad media: {age_mean}")

age_data["Desv"] = age_data["Age"] - age_mean
print(f"Desviación media absoluta Age manual: {age_data['Desv'].abs().mean()}")
```

```python
# Desviación media absoluta de columna Age (Pyhton)
print(f"Desviación media absoluta Age python: {age_data['Age'].mad()}")
```

Obtener la desviación Media Absoluta de la columna **Fare** de manera manual, posterirmente obtenerla con la funcion de Phyton y comparar resultados

```python
# Desviación media absoluta de columna Fare (manual)
fare_data = df[["Fare"]].copy()
print(fare_data.shape)

fare_mean = fare_data.mean()[0]
print(f"Tarifa media: {fare_mean}")

fare_data["Desv"] = fare_data["Fare"] - fare_mean
print(f"Desviación media absoluta Fare Manual: {fare_data['Desv'].abs().mean()}")
```

```python
# Desviación media absoluta de columna Fare (Pyhton)
print(f"Desviación media absoluta Fare python: {fare_data['Fare'].mad()}")
```

Crea una gráfica para observar de manera visual como se esta comportando la desviación contra los demas datos en la columna **Age**



```python
age_ax = age_data["Age"].plot(style="o")
age_ax.figure.set_size_inches(14,8)


for i in range(age_data.shape[0]):
    age_ax.vlines(x = i,
                  ymin = age_mean,
                  ymax = age_data["Age"][i],
                  linestyles = "dashed",
                  color= "gray"
                 )
    
age_ax.hlines(y=age_mean,
              xmin=0,
              xmax=age_data.shape[0]-1,
              colors = "red",
              linewidth=2
             )
```

Crea una gráfica para observar de manera visual como se esta comportando la desviación contra los demas datos en la columna **Fare**

```python
fare_ax = fare_data["Fare"].plot(style='o')
fare_ax.figure.set_size_inches(14,8)
for i in range(fare_data.shape[0]):
    fare_ax.vlines(x = i,
              ymin = fare_mean,
              ymax = fare_data["Fare"][i],
              linestyles = "dashed",
              color = "gray"
             )

fare_ax.hlines(y = fare_mean,
               xmin=0,
               xmax = fare_data.shape[0] -1,
               colors = "red",
               linewidth = 2
              )
```

## <font color=green>Varianza</font>
***


Obtener la varianza de la columna **Age** de manera manual, posterirmente obtenerla con la funcion de Phyton y comparar resultados

```python
# Varianza de columna Age (manual)
age_data["Desv^2"] = age_data["Desv"].pow(2)
age_data
age_var = age_data["Desv^2"].sum() / (len(age_data) -1)
print(f"Varianza Age Manual: {age_var}")
```

```python
# Varianza de columna Age (python)
print(f"Varianza Age python: {age_data['Age'].var()}")
```

Obtener la varianza de la columna **Fare** de manera manual, posterirmente obtenerla con la funcion de Phyton y comparar resultados

```python
# Varianza de columna Fare (manual)
fare_data["Desv^2"] = fare_data["Desv"].pow(2)
fare_data
fare_var = fare_data["Desv^2"].sum() / (len(fare_data) -1)
print(f"Varianza Fare manual: {fare_var}")
```

```python
# Varianza de columna Age (python)
print(f"Varianza Fare python: {fare_data['Fare'].var()}")
```
