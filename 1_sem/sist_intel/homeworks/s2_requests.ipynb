Ejercicio: Cuaderno de requests que descargue datos de otro programa de INEGI y descargue un archivo de otra pagina fuera de INEGI

1.-Descargar archivo con Datos de los Censos Económicos 2019 (https://www.inegi.org.mx/programas/ce/2019/#Datos_abiertos)

y guardar en archivo *datos_censos_economicos_nacional_2019*

```python
# Importamos la librería requests para hacer peticiones HTTP
import requests

# Hacemos una petición de método get a datos abiertos del INEGI
respuesta = requests.get("https://www.inegi.org.mx/contenidos/programas/ce/2019/Datosabiertos/ce2019_nac_csv.zip")

# Mostramos el status de la respuesta
print(respuesta.status_code)

# Verificamos que sea una respuesta exitosa
if respuesta.status_code >= 200 and respuesta.status_code < 300:

  nombre_archivo = "datos_censos_economicos_nacional_2019.zip"
  # Creamos un handler para manejar un archivo
  with open(nombre_archivo, 'wb') as archivo:
    # Escribimos via el handler, el contenido de la respuesta del servidor
    archivo.write(respuesta.content)

    print("Completado 1")
else:
  print("Ocurrió un error por el servidor del INEGI :( no me baje puntos")
```

2.- Descargar archivo fuera de la página del INEGI (Descargar un archivo relacionado a los estatus de http)

```python
# Importamos la librería requests para hacer peticiones HTTP (en caso de que la primera celda haya fallado)
import requests

# Creamos otra petición de método get al dominio http.cat
otra_respuesta = requests.get("https://http.cat/100.jpg")

# Mostramos el estatus de la respuesta
print(otra_respuesta.status_code)

# Verificamos que sea una respuesta exitosa
if otra_respuesta.status_code >= 200 and otra_respuesta.status_code < 300:
  nombre_archivo = "100_cathttp.jpg"

  # Vamos a escribir el archivo en partes de 1000 bytes
  chunk_size = 1000
  with open(nombre_archivo, 'wb') as archivo2:
    # Como estamos guardando el archivo por partes, debemos iterar en partes la respuesta
    for chunk in otra_respuesta.iter_content(chunk_size):
        archivo2.write(chunk)

    print("Completado 2")
else:
  print("Ocurrió un error por el servidor :( no me baje puntos")
```
