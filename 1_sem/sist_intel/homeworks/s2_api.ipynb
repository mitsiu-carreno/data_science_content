```python
TOKEN_CORTESIA_INEGI='96fbd1bf-21e6-28e3-6e64-2b15999d2c89'
```

Ignoro si estos tokens tienen expiración, pero por su tipo de uso, no creo. También creo que no es muy seguro tenerlos en la URL.

```python
url=f'https://www.inegi.org.mx/app/api/indicadores/desarrolladores/jsonxml/INDICATOR/1006000039/es/0700/false/BISE/2.0/{TOKEN_CORTESIA_INEGI}?type=json'
# Realizamos petición a API de INEGI
respuesta= requests.get(url)

# Veficiamos que respuesta es exitosa
if respuesta.status_code >= 200 and respuesta.status_code < 300:
  # Convertimos nuestro string en un diccionario de python valido
  contenido = json.loads(respuesta.content)

  # Entramos a los atributos jugosos
  observaciones = contenido["Series"][0]["OBSERVATIONS"]
  # Recorremos nuestro arreglo de observaciones
  print(f'Año\t- Observaciones de Accidentes de tránsito terrestre en zonas urbanas y suburbanas')
  for observacion in observaciones:
    print(f'{observacion["TIME_PERIOD"]}\t- {observacion["OBS_VALUE"]}')
else:
  print("Ocurrió un error por el servidor :( no me baje puntos")
```

# 2: API SunriseSunset.io

```python
#Ejercicio 2:

url=f'https://api.sunrisesunset.io/json?lat=21.880179&lng=-102.3037028&timezone=CST&date=today'
# Realizamos petición a API de INEGI
respuesta= requests.get(url)

# Veficiamos que respuesta es exitosa
if respuesta.status_code >= 200 and respuesta.status_code < 300:
  # Convertimos nuestro string en un diccionario de python valido
  contenido = json.loads(respuesta.content)

  # Nuestra respuesta tiene un diccionario en results, iteramos cada elemento del diccionario
  for key, val in contenido["results"].items():
    print(f"{key} - {val}")

else:
  print("Ocurrió un error por el servidor :( no me baje puntos")

```
