```python
from bs4 import BeautifulSoup
import requests

site_domain = "https://en.wikipedia.org"

random_site = requests.get(site_domain + "/wiki/Special:Random")

def GetFullLink(link):
    '''
    Description: 
        Get the full url given a relative or absolute URL
    Input:
        A link (the href value)
    Output:
        A url starting by protocol, domain and resources
    '''
    if link.startswith("http"):
        return link;
    if link.startswith("//"):
        return "https:" + link
    if link.startswith("#") or link.startswith("/"):
        return site_domain + link;
    else:
        return site_domain + "/" + link

    
random_soup = BeautifulSoup(random_site.text, 'html.parser')

# Debug
#print(random_soup.prettify())
print(random_soup.title)

# Get all html a tags
links = random_soup.find_all('a')

# We must filter only links containing wiki
# I decided to use a lambda to learn the sintax (basically an anonymous function)
# In the function we check if our a tag has href and if wiki exists in said href
links = list(filter(lambda link: link.has_attr('href') and 'wiki' in link.get('href'), links))


tot_links = len(links)

for i, link in enumerate(links):
    
    full_link = GetFullLink(link.get('href'))

    print(f"Listing images found in link {i+1}/{tot_links}: {full_link}")
    # Get linked site
    linked_site = requests.get(full_link)
    
    if linked_site.ok == False:
        print("Site not reachable")
    
    linked_soup = BeautifulSoup(linked_site.text, 'html.parser')
    
    # Get all html img tags
    images = linked_soup.find_all('img')
    for image in images:
        if not image.has_attr('src'):  # img tag without src
            print("HTML img tag doesn't source :(")
            continue # Skip to next image
        print("\t"+GetFullLink(image.get('src')))
    print("\n")
    
```
