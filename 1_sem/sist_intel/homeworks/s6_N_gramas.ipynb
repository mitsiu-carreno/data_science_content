```python
# Importamos las librer√≠as necesarias
import requests
import nltk
nltk.download('stopwords')
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer


url_libro = "https://www.gutenberg.org/files/1938/1938-0.txt"

res = requests.get(url_libro)
res.encoding = res.apparent_encoding
documento = res.text

stop_words = nltk.corpus.stopwords.words('english')

# Tambi√©n se puede usar ngrams de nltk 
#tokens = nltk.ngrams(documento.split(), 2)

# Agregamos el parametro ngram_range=(2,2) para generar 2-gramas
vectorizador = CountVectorizer(ngram_range = (2, 2), input="content", analyzer="word", stop_words=stop_words)
matriz_frecuencias = vectorizador.fit_transform([documento])

tabla_frecuencias = pd.DataFrame(
    np.transpose(matriz_frecuencias.toarray()),
    index=vectorizador.get_feature_names_out(),
    columns = ["conteo"]
)

tabla_frecuencias.sort_values("conteo", ascending=False).head(50)
```

Un caso de uso para los n-gramas en especifico con los 2-gramas puede ser un modelo de texto predictivo como el que tienen los celulares de autocompletado, que toman la √∫tima palabra escrita para intentar predecir la siguiente.
