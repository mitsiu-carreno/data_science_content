```python
## Mitsiu Alejandro Carreo Sarabia - E23S-18014
### Ejercicio 7 - Componentes principales

import numpy as np
from sklearn.decomposition import PCA
from sklearn.datasets import load_wine
import seaborn as sns
import pandas as pd

# Cargamos el dataset
data = load_wine()
x = data.data

#print(data.DESCR)

# Cargamos los datos en un dataframe
df = pd.DataFrame(data = data.data, columns=data["feature_names"])

# Podemos grficar las relaciones entre distintas variables
sns.pairplot(df[data['feature_names'][:5]])

# Creamos un modelo de analisis de componentes
pca = PCA()
x_pca = pca.fit_transform(x)

print("Varianza por cada componente principal")
print(pca.explained_variance_ratio_)

pca_df = pd.DataFrame(data=x_pca)

# El nivel de variabilidad con dos variables llega a 99.98
print(np.cumsum(pca.explained_variance_ratio_))

# Generamos un nuevo modelo de componentes principales con nicamente dos variables
pca_2 = PCA(n_components=2)
x_pca_2 = pca_2.fit_transform(x)

# Creamos un nuevo dataframe con los valores restringidos
pca_2_df = pd.DataFrame(data = x_pca_2)
print(pca_2_df)

sns.pairplot(pca_2_df)
```
