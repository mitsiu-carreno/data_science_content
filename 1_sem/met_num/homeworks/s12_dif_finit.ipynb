¿Cómo las podemos implementar en Python?

R: Las diferencias finitas se pueden aplicar de manera central o lateral (las laterales a su vez se dividen en progresiva o regresiva), nos permiten obtener las derivadas de funciones que son muy complejas o no existen.

También se pueden usar para calcular la integral, al sumar las áreas de cada intervalo, pero para ello es necesario tener intervalos muy pequeños, que se acerquen más al área que se esta midiendo.

Usualmente en los fenomenos naturales, interactuan distintos aspectos en un mismo fenomeno, por lo que calcularlo de manera exacta, involucraría un número muy grande de variables con una precisión más allña de la capacidad actual de las computadoras, es por esto que calcularlas mediante diferencias finitas, permite, realizar los calculos en tiempo, procesamiento y espacio finitos a costo de perder precisión tanto en calculos como en el resultado.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
coeficientes = np.array([-5,0,2,7,5])
coeficientes
```

```python
x = np.linspace(-1.5, 1.5, 14)  

y = np.polyval(coeficientes, x)


y_central = (y[:-1] + y[1:])/2

plt.plot(x, y, label="$f(x) = -5x^4+3x^2+1x+5$")
plt.step(x, y, c="g")
plt.scatter(x, y, label="Lateral Regresiva")
plt.scatter(x[:-1], y[1:], label="Lateral Progresiva")
plt.scatter(x[:-1], y_central, label="Central", c="r")
plt.xlabel('x')
plt.ylabel('y')
plt.title('$f(x) = -5x^4+3x^2+1x+5$')
plt.grid(True)
plt.legend(loc="lower right")
plt.show()
```
