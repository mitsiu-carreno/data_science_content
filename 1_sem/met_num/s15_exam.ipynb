### 1.1 Identifica si son invertibles, de caso contrario no continues con los siguientes pasos

```python
def IsInvertible(mat):
    '''
    Descrip: Return whether a matrix is invertible
    Input:
        mat (nxm): Matrix to be evaluated
    Output:
        bool
    '''
    n,m = mat.shape
    if n != m:
        return False
    
    return np.linalg.det(mat) != 0
```

```python
print(f"La matriz A es invertible: {IsInvertible(A)}")
print(f"La matriz B es invertible: {IsInvertible(B)}")
print(f"La matriz C es invertible: {IsInvertible(C)}")
```

Explicación: Las matrices invertibles o no singulares tienen un determinante distinto de cero, cabe mencionar que la matriz B se acerca a cero, por lo que es posible que sea una matriz mal condicionada

### 1.2. Muestra los pasos intermedios para obtener la matriz mediante Gauss-Jordan

```python
def ImprimirSistema(m_coef, m_inver):
    for fila, fila_inv in zip(m_coef, m_inver):
        fila_formato = "\t".join([f"{x: 4.2f}" for x in fila])
        fila_inv_formato = "\t".join([f"{x: 4.2f}" for x in fila_inv])
        print(f"{fila_formato}\t|\t{fila_inv_formato}")
        
def InvGaussJordan(m_coef):
    matriz = m_coef.astype(float)
    n_filas, n_cols = matriz.shape
    inversa = np.eye(n_filas)
    
    for columna in range(n_cols):
        for fila in range(n_filas):
            if fila == columna:
                print(f"R{fila} --> R{fila}*1/{matriz[fila,columna]}")
                factor = 1/matriz[fila,columna]
                matriz[fila,:] = matriz[fila,:] * factor
                inversa[fila,:] = inversa[fila,:] * factor
            else:
                factor =- matriz[fila, columna] / matriz[columna,columna]
                #print(f"factor de {factor} ")
                print(f"R{fila} --> {factor}*R{columna} + R{fila}") 
                matriz[fila,:] = matriz[columna,:] * factor + matriz[fila,:]
                inversa[fila,:] = inversa[columna,:] * factor + inversa[fila,:]
            ImprimirSistema(matriz, inversa)
            print("\n")
    return inversa

print("Resolviendo inversa de A por Gauss-Jordan")
A_gj_inv = InvGaussJordan(A)
print("\nInversa de A:")
print(A_gj_inv)

print("\n\n==========================================\n\n")

print("Resolviendo inversa de B por Gauss-Jordan")
B_gj_inv = InvGaussJordan(B)
print("\nInversa de B:")
print(B_gj_inv)
```

Explicación: Para obtener la inversa de una matriz $A$, se crea un sistema en el que se une $A|I$ donde $I$ es la matriz identidad del mismo tamaño que la matriz $A$, se resuelve el sistema mediante Gauss-Jordan y se obtiene $I|B$ donde $B$ es la matriz identidad correspondiente a la matriz $A$

### 1.3 Obtén la inversa mediante librerías de Python

```python
A_np_inv = np.linalg.inv(A)
print("Inversa de A (numpy)")
print(A_np_inv)

B_np_inv = np.linalg.inv(B)
print("Inversa de B (numpy)")
print(B_np_inv)
```

Explicación: Numpy ofrece el método [inv](https://numpy.org/doc/stable/reference/generated/numpy.linalg.inv.html) del módulo linalg que resuelve la inversa de una matriz.

### 1.4 Comprueba que la matriz GJ sea realmente la inversa

```python
print("Comprobando la inversa de A (Gauss-Jordan)")
print(np.allclose(
    np.matmul(A, A_gj_inv),
    np.eye(A.shape[0])
))

print("Comprobando la inversa de B (Gauss-Jordan)")
print(np.allclose(
    np.matmul(B, B_gj_inv),
    np.eye(B.shape[0])
))
```

Explicación: Para comprobar una inversa, basta con multiplicar la matriz original con su inversa y el resultado debe ser la matriz identidad. En este caso la matriz B no conicide, esto (y el hecho de que la inversa mediante numpy sea diferente) confirma que B es una matriz mal condicionada y por ello produce estos errores.

```python
print(f"La matriz A es invertible: {IsInvertible(A)}")

A_inv = np.linalg.inv(A)
x = np.matmul(A_inv, b)

print("Los valores de x son:")
print(x)

print("Confirmando respuestas")
np.allclose(
    np.matmul(A,x),
    b
)
```

Explicación: La matriz de coeficientes A y el vector de resultados b se relacionan con nuestro vector de incognitas de la siguiente manera $Ax = b$ si queremos despejar x, debemos pasar A al lado contrario con la operación contraria (multiplicar-> dividir) pero al ser una matriz, no se puede hacer directamente. Pero, por ser una matriz, tiene una inversa que cumple con la siguiente caracteristica $A^{-1} = 1/A$ por ello podemos convertir $Ax = b$ en $x = A^{-1}b$ obtenemos la inversa de A, la multiplicamos por b y obtenemos los valores de x

### 4.1 Calcula las diferencias hacia adelante de $f(x) = x^3+2x^2+4x-2  $ en el intervalo de [-5, 5] con 20 puntos.

```python
import matplotlib.pyplot as plt
```

```python
# Ingresamos la función y intervalo dados
coef = np.array([1,2,4,-2])
x = np.linspace(-5,5,20)

# Evaluamos para obtener los valores de y
y = np.polyval(coef, x)

# Gráficamos la función
plt.plot(x, y)
plt.title("$f(x) = x^3+2x^2+4x-2$")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
# Calculamos las diferencias finitas adelante
forward_diff = np.diff(y)/np.diff(x)
x_diff = x[:-1:]
exact_sol = np.polyval(np.polyder(coef), x_diff)

# Gráficamos la función
plt.plot(x_diff, exact_sol, label="Derivada exacta")
plt.step(x_diff, forward_diff, "--", )
plt.scatter(x_diff, forward_diff, c="r", label="Diferencias finitas adelante")
plt.title("$f'(x) = 3x^2+4x+4$")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```

### 4.2 Calcula las diferencias centrales de $f(x) = \frac{1} {2+sen(x)} $ en los puntos $1.1,1.2,1.3,...,3.0$

```python
# Ingresamos la función 
def f(x):
    return 1/(2+np.sin(x))

# Ingresamos los intervalo dados
x = np.arange(1.1,3,.1)

# Evaluamos para obtener los valores de y
y = f(x)

# Gráficamos la función
plt.plot(x, y)
plt.title(r"$f(x) = \frac{1}{2+sen(x)}$")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
def grad_f(x):
    return - (np.cos(x))/((2+np.sin(x))**2)

# Calculamos las diferencias finitas centrales
central_diff = (y[2:]-y[:-2]) / (x[2:]-x[:-2])
x_diff = x[1:-1]
exact_sol = grad_f(x_diff)

# Gráficamos la función
plt.plot(x_diff, exact_sol, label="Derivada exacta")
plt.step(x_diff, central_diff, "--", )
plt.scatter(x_diff, central_diff, c="r", label="Diferencias finitas centrales")
plt.title(r"$f'(x) = -\frac{cos(x)}{(2+sen(x))^2}$")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```

# Ejercicio 6

Encuentra el punto mínimo de la función a continuación.

explica como elegir el punto inicial y el tamaño de paso


```python
import matplotlib.pyplot as plt
```

```python
def funcion_e6(x,y):
  return np.polyval(np.array([ 1,  0,  -2]), x) + np.polyval(np.array([ 1,  -3,  2]), y) +  2*np.sin(10*y) + np.cos(10*x)
```

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x,y)

Z = funcion_e6(X, Y)
plt.title("Curvas de nivel $f(x, y) = x^2+y^2-3y+cos(10x)+2sin(10y)$")
plt.xlabel("x")
plt.ylabel("y")
plt.contour(X, Y, Z, levels=40)

plt.show()
```

```python
def grad_f(x, y):
    return np.array([2*(x-5*np.sin(10*x)), 2*y+10*np.cos(10*y)-3])

def decenso_grad(grad, tam_paso, x0, iteraciones):
    pasos = [x0.copy()]
    xi = x0
    for i in range(iteraciones):
        derivada = grad(*xi)
        xi -= tam_paso * derivada
        pasos.append(xi.copy())
    return pasos

pasos = decenso_grad(grad_f, 0.12, np.array([-10,10], dtype=float), 10)
pasos_x = [x[0] for x in pasos]
pasos_y = [x[1] for x in pasos]


plt.title("Curvas de nivel $f(x, y) = x^2+y^2-3y+cos(10x)+2sin(10y)$")
plt.xlabel("x")
plt.ylabel("y")
plt.contour(X, Y, Z, levels=40)
plt.plot(pasos_x,pasos_y,color="blue")
plt.scatter(pasos_x,pasos_y,color="red")

plt.show()
```

Explicación: El descenso de gradiente es una función que nos permite maximizar o minimizar ciertos valores, por ello es útil para encontrar el punto minimo de una función, internamente el proceso consiste en calcular la derivada de cierto punto, y moverse en dirección contraria a dicho gradiente,  buscando el punto minimo de la función. Para la elección del punto inicial y el tamaño de paso, es importante notar, que el punto de paso debe ser lo suficientemente grande como para "salir" de minimos locales pero lo sucientemente "pequeño" para no "saltar" el minimo global; el punto inicial es menos crítico y se puede llegar incluso a considerar un punto inicial aleatorio.

$$ 2\sin\left(x^{2}\right)+\sin\left(x\cos\left(x\right)+2\right)+1.3 $$
Intervalo de x [-2, 2.17]


```python
def f(x):
    return 2*np.sin(x**2)+np.sin(x*np.cos(x)+2)+1.3

def integ_trapecio(func, x):
    y = func(x)
    dif_x = x[1:] - x[:-1]
    trapecios = (y[1:] + y[:-1])/2
    integral = sum(trapecios*dif_x)
    
    x_plot = np.linspace(x[0], x[-1], 101)
    y_plot = func(x_plot)
    
    plt.plot(x_plot, y_plot, color="black", alpha=0.3)
    plt.fill_between(x, y, color="gold", alpha=0.7)
    plt.scatter(x,y, c="r", s=5)
    
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title(f"Terrorifica Boa deborando elefante \n o en terminos que entiendan los adultos $2sin(x^2)+sin(x cos(x)+2)+1.3$")
    plt.grid(True)
    plt.annotate(f"Área de boa = {integral}", xy=(0.1,0.2),xycoords='axes fraction',
             fontsize=14)
    plt.show()
    
    return integral
    
area = integ_trapecio(f, np.linspace(-2, 2.17, 20))
print(f"Area {area}")

```
