```python
personas = []
for i in range(1,11):
#for i in range(1,3):
    #persona = {"nombre":"", "edad":""}
    persona = {}
    persona["nombre"] = input(f"\nIngresa el nombre de la persona {i}: ")
    while(True):
        try:
            persona["edad"] = int(input(f"Ingresa la edad de la persona {i} ({persona['nombre']}): "))
            break
        except:
            print("Edad invalida, ingrese un número entero\n")
    personas.append(persona)

for persona in personas:
    if persona["edad"] >= 18:
        print(f"\n{persona['nombre']} es mayor de edad")
        if persona["edad"] >= 60:
            print(f"y {persona['nombre']} puede obtener descuento de 3ra edad")

def orderByAge(persona):
  return persona["edad"]

personas.sort(reverse=True, key=orderByAge)

print("\nMostrando personas por orden de edad de mayor a menor\n");
for persona in personas:
    print(f"{persona['edad']} - {persona['nombre']}")
```
