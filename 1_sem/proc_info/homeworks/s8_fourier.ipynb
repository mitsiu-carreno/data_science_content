5- Generate two sinals signal 1 is a sine wave with 5 Hz, amplitude 3 and phase shift 3; signal 2 is a sine wave with 2 Hz, amplitude 2 and phase shift -2 Plot the signal for 2 seconds

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-poster')
%matplotlib inline

# sampling rate
sr = 100.0
# sampling interval
ts = 1.0/sr
t = np.arange(0,2, ts)

plt.figure(figsize=(6,10))
plt.subplot(311)
plt.ylabel("Amplitude")

##### Signal 1
freq = 5
ampl = 3
phase = 3
y1 = ampl* np.sin(2*np.pi * freq * t + phase)
plt.plot(t, y1, 'b')

##### Signal 2
freq = 2
ampl = 2
phase = -2
y2 = ampl * np.sin(2*np.pi * freq * t + phase)
plt.subplot(312)
plt.plot(t, y2, 'b')
plt.ylabel("Amplitude")


##### Signals combined
y1 +=  y2
plt.subplot(313)
plt.plot(t, y1, 'b')
plt.ylabel("Amplitude")
plt.xlabel("Time (s)")
plt.show()
```

9- We implement DFT previously, can you implement the inverse Discrete Fourier Transform in Python similarly?

```python
def IDFT(x):
    """
    Function to calculate the 
    inverse discrete Fourier transform
    of a 1-D real-valued signal x
    """
    N = len(x)
    n = np.arange(N)
    k = n.reshape((N,1))
    e = np.exp((2j*np.pi * k * n)/N)
    X = np.dot(e, x)
    return 1/N * X
```

10- Use the DFT function and inverse DFT we implemented and generate the amplitude spectrum for the signal you generated in problem 5. Normalize the DFT amplitude to get the correct corresponding time domain and amplitude

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-poster')
%matplotlib inline

def DFT(x):
    """
    Function to calculate the 
    discrete Fourier Transform
    of a 1D real-valued signal x
    """
    
    N = len(x)
    n = np.arange(N)
    k = n.reshape((N,1))
    #e = np.exp(-2j * np.pi * k * n / N)
    e = np.exp((-2j * np.pi * k * n) / N)
    
    X = np.dot(e, x) # Dot product of two arrays
    
    return X

# sampling rate
sr = 100.0
# sampling interval
ts = 1.0/sr
t = np.arange(0,2, ts)

##### Signal 1
freq = 5
ampl = 3
phase = 3
y = ampl* np.sin(2*np.pi * freq * t + phase)

##### Signal 2
freq = 2
ampl = 2
phase = -2
y += ampl * np.sin(2*np.pi * freq * t + phase)

##### Graph signal
plt.figure(figsize=(6,6))
plt.subplot(211)
plt.plot(t, y, 'b')
plt.ylabel("Amplitude")
plt.xlabel("Time (s)")

## Discrete Fourier Transform
X = DFT(y)

# Calculate the frequency
N = len(y)
n = np.arange(N)
T = N/sr
freq = n/T
plt.subplot(212)
plt.stem(freq, abs(X), 'b', \
        markerfmt=" ", basefmt="-b")
plt.xlabel("Freq (Hz)")
plt.ylabel("DFT Amplitude |X(freq)|")
plt.show()
```

11- Can you describe the tricks used in FFT to make the computation faster?    
R- Como se puede apreciar en la grafica de frequencias del ejercicio anterior, la transformación de Fourier genera senos simetricos, en realidad con DFT estamos realizando más computaciones de las que realmente son absolutamente necesarias, FFT toma una aproximación de Dividir y conquistar en la que el problema se divide en problemas similares pero más pequeños, con ello se termina con un muchos problemas sencillos en lugar de un problema complejo, ello para la computadora es más sencillo de procesar. Para generar los problemas más sencillos, se emplea una técnica llamada "Recursividad" en la cuál el mismo algoritmo se invoca n veces pero cada vez con inputs más pequeños.
