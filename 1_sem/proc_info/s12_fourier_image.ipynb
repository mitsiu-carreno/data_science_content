```python
from skimage import io
from skimage.color import rgb2gray

img = io.imread('s12_fourier_image_nature.png')

gray_img = rgb2gray(img[:,:,:3])
io.imshow(gray_img)
```

```python
# Unify format for gray values from 0 to 255 (gray images in python can have float values)
from skimage.util import img_as_ubyte
gray_img = img_as_ubyte(rgb2gray(img[:,:,:3]))
io.imshow(gray_img)
```

```python
# Fast fourier transform
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

f = np.fft.fft2(gray_img)
f_s = np.fft.fftshift(f)
plt.figure(num=None, figsize=(10,8), dpi=80)
plt.imshow(np.log(abs(f_s)), cmap='gray')
plt.show()
```
