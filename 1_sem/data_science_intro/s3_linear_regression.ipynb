```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def lineplot(a, b):
    '''
    inputs:
        a: [pendiente]
        b: [ordenada del origen]
    output:
        Gr치fica de una recta
    '''
    if(len(a) != len(b)):
        print("Vector de pendientes debe tener la misma longitud que vector de ordenadas")
        return 
    
    for i in range(len(a)):
        x=list(range(1,10))
        y=a[i]*np.array(x) + b[i]
    
        plt.plot(x, y, label=f'y={a[i]}x+{b[i]}')
        plt.legend()
    plt.show()
    
```

```python
lineplot([1,2,3],[3,3,3])
```

# M칤nimos cuadrados ordinarios
Equation 1  ^Bo = y(prom) - ^B1x(prom)

Equation 2  B1 = \frac(SUM(xi-x(prom))(yi-y(prom)))(SUM(xi-x(prom))^2)

```python
x = data["distancia"].copy()
y = data["precio"].copy()


b1 = sum((x-x.mean())*(y-y.mean()))/sum((x-x.mean())**2)

b1 = round(b1, 2)

b0 = y.mean()-b1*x.mean()

print(f'Nuestra ecuaci칩n es Precio={b1}*distancia+{b0}')
```

```python
precio = b1*5.5+b0
print(precio)
```

# Regresi칩n con paqueter칤a

```python
from sklearn.linear_model import LinearRegression
```

```python
x=np.array(data["distancia"]).reshape(20,1)
y=data["precio"]

reg = LinearRegression().fit(x,y)

print(reg.coef_)

print(reg.intercept_)

print(reg.predict(np.array([[5.5]])))
```
