```python
import nltk
from bs4 import BeautifulSoup
import requests
```

```python
#url="https://presidente.gob.mx/23-06-23-version-estenografica-de-la-conferencia-de-prensa-matutina-del-presidente-andres-manuel-lopez-obrador-desde-chiapas/"
url="https://www.gob.mx/presidencia/articulos/discurso-de-andres-manuel-lopez-obrador-presidente-de-los-estados-unidos-mexicanos?idiom=es"


resp = requests.get(url)
txt = resp.text
```

```python
txt
```

```python
htmlParse = BeautifulSoup(txt, "html.parser")
corpus = []

for parag in htmlParse.find_all("p"):
    print(parag.get_text())
    corpus.extend(parag)
```

```python
len(corpus)
```

```python
pattern = r'''(?x)                 # set flag to allow verbose regexps
              (?:[A-Z]\.)+         # abbreviations, e.g. U.S.A.
              | \w+(?:-\w+)*       # words with optional internal hyphens
              | \$?\d+(?:\.\d+)?%? # currency and percentages, e.g. $12.40, 82%
              | \.\.\.             # ellipsis
              | [][.,;"'?():-_`]   # these are separate tokens; includes ], [
'''
nltk.regexp_tokenize(corpus[1], pattern)
```

```python
tokens = []

for row in corpus:
    try:
        tokens.extend(nltk.regexp_tokenize(row, pattern))
    except:
        print(row)
        continue
    
print(tokens)
```

```python
fdist = nltk.FreqDist(tokens)

fdist.most_common(20)
```
