```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Crear centroides reales
c1 = [-10,-110]
c2 = [-10,10]
c3 = [10,-10]
c4 = [10,10]

c = np.array([c1, c2, c3, c4])
c
np.amin(c, axis=0)
#np.amin(c)
```

```python
#crear puntos nuevos
from random import randrange

px = np.array([])
py = np.array([])

# Populate points x
for centr in c:
    print("1")
    for e in range(randrange(10, 20)):
        px = np.append(px, centr[0] + randrange(5) if centr[0] > 0 else centr[0] - randrange(5))
        py = np.append(py, centr[1] + randrange(5) if centr[1] > 0 else centr[1] - randrange(5))
```

```python
# https://stackoverflow.com/questions/13430231/how-i-can-get-cartesian-coordinate-system-in-matplotlib
import numpy as np                 # v 1.19.2
import matplotlib.pyplot as plt    # v 3.3.2

def DrawPlot(px, py):
    # Select length of axes and the space between tick labels
    #np.amax(c)
    #np.amin(c)
    xmin, xmax, ymin, ymax = -16, 16, -16, 16
    ticks_frequency = 1

    # Plot points
    fig, ax = plt.subplots(figsize=(10, 10))
    ax.scatter(px, py, c="b")
    ax.scatter(1, 1, c="r")

    # Set identical scales for both axes
    ax.set(xlim=(xmin-1, xmax+1), ylim=(ymin-1, ymax+1), aspect='equal')

    # Set bottom and left spines as x and y axes of coordinate system
    ax.spines['bottom'].set_position('zero')
    ax.spines['left'].set_position('zero')

    # Remove top and right spines
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    # Create 'x' and 'y' labels placed at the end of the axes
    ax.set_xlabel('x', size=14, labelpad=-24, x=1.03)
    ax.set_ylabel('y', size=14, labelpad=-21, y=1.02, rotation=0)

    # Create custom major ticks to determine position of tick labels
    x_ticks = np.arange(xmin, xmax+1, ticks_frequency)
    y_ticks = np.arange(ymin, ymax+1, ticks_frequency)
    ax.set_xticks(x_ticks[x_ticks != 0])
    ax.set_yticks(y_ticks[y_ticks != 0])

    # Draw major and minor grid lines
    ax.grid(which='both', color='grey', linewidth=1, linestyle='-', alpha=0.2)

    # Draw arrows
    arrow_fmt = dict(markersize=4, color='black', clip_on=False)
    ax.plot((1), (0), marker='>', transform=ax.get_yaxis_transform(), **arrow_fmt)
    ax.plot((0), (1), marker='^', transform=ax.get_xaxis_transform(), **arrow_fmt)

    plt.show()
```

```python
DrawPlot(px, py)
```
